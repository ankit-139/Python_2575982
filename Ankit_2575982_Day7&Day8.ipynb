# 1. Write a Python program to find a target values in a list using linear search with following steps:

a. Initialize the list to store the input elements.

b. Initialize found-False.

C. Enter the item to be searched (match_item).

d. For each element in the list

        (i). if match_item = value

        (ii). return match_item's position.

e. If the match item is not in the list, display an error message that the item is not found in the list.

```python
def linear_search(input_list, match_item):
    found = False
    position = -1

    for index, value in enumerate(input_list):
        if value == match_item:
            found = True
            position = index
            break

    return found, position

# Initialize the list
input_list = [100, 200, 300, 400, 500]

# Initialize found to False
found = False

# Enter the item to be searched (match_item)
match_item = int(input("Enter the item to be searched: "))

# Perform linear search
found, position = linear_search(input_list, match_item)

# Display the result
if found:
    print(f"Item {match_item} found at position {position}.")
else:
    print(f"Item {match_item} not found in the list.")
```

# 2. Write a Python program to implement binary search to find the target values from the list:

a. Create a separate function to do binary search.

b. Get the number of inputs from the user.

C. Store the inputs individually in a list.

d. In binary search function at first sort the list in order to start the search from middle of the list.

e. Compare the middle element to right and left elements to search target element.

f. If greater, move to right of list or else move to another side of the list.

g. Print the result along with the position of the element.

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            return mid  # Target found
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1  # Target not found

def main():
    # Get the number of inputs from the user
    num_inputs = int(input("Enter the number of elements: "))

    # Store the inputs individually in a list
    input_list = []
    for i in range(num_inputs):
        element = int(input(f"Enter element {i + 1}: "))
        input_list.append(element)

    # Sort the list
    input_list.sort()

    # Get the target value from the user
    target = int(input("Enter the target value to search: "))

    # Perform binary search
    position = binary_search(input_list, target)

    # Print the result
    if position != -1:
        print(f"Target value {target} found at position {position}.")
    else:
        print(f"Target value {target} not found in the list.")

if __name__ == "__main__":
    main()
```

# 3. Write a Python program for sorting a list of elements using selection sort algorithm:

a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted list.

c. In the second iteration, find the smallest element in the unsorted list and place it in the correct position by comparing with the element in the sorted list.

d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position by comparing with the elements in the sorted list.

e. This process continues till the unsorted list becomes empty.

f. Display the sorted list

```python
def selection_sort(arr):
    sorted_list = []  # Initialize an empty sorted list

    while arr:  # Loop until the unsorted list becomes empty
        min_index = 0

        # Find the index of the minimum element in the unsorted list
        for i in range(1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i

        # Remove the minimum element from the unsorted list and add it to the sorted list
        sorted_list.append(arr.pop(min_index))

    return sorted_list

def main():
    # Input list
    input_list = [46, 52, 21, 30, 10, 11, 63]

    # Perform selection sort
    sorted_list = selection_sort(input_list)

    # Display the sorted list
    print("Sorted list:", sorted_list)

if __name__ == "__main__":
    main()
```

# 4.Write a Python program for sorting a list of elements using insertion sort algorithm:

a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

b. In the first iteration, take the first element in the unsorted list and insert it in Sorted list.

c.In the second iteration, take the second element in the given list and compare with the element in the sorted sub list and place it in the correct position.

d. In the third iteration, take the third element in the given list and compare with the elements in the sorted sub list and place the elements in the correct position.

e. This process continues until the last element is inserted in the sorted sub list.

f. Display the sorted elements.

```python
def insertion_sort(arr):
    # Iterate through each element in the unsorted list
    for i in range(1, len(arr)):
        # Store the current element to be inserted
        current_element = arr[i]
        j = i - 1
        
        # Move elements of the sorted sub-list that are greater than the current element
        # to the right, shifting them one position ahead
        while j >= 0 and current_element < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        
        # Insert the current element in its correct position
        arr[j + 1] = current_element

# Input list of elements
unsorted_list = [102, 1, 3, 50, 66]

# Call the insertion_sort function to sort the list
insertion_sort(unsorted_list)

# Display the sorted elements
print("Sorted List:")
for element in unsorted_list:
    print(element, end=" ")
```

# 5. Write a Python program that performs merge sort on a list of numbers:

a. Divide: If the given array has zero or one element, return.

    i. Otherwise

    ii. Divide the input list in to two halves each containing half of the elements. i.e. left half and right half.

b. Conquer: Recursively sort the two lists (left half and right half).

    i. Call the merge sort on left half.

    ii. Call the merge sort on right half.

c. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return
    
    # Divide the input list into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    
    # Recursively sort the left and right halves
    merge_sort(left_half)
    merge_sort(right_half)
    
    # Merge the two sorted halves back into a single sorted list
    merge(arr, left_half, right_half)

def merge(arr, left_half, right_half):
    i = j = k = 0
    
    # Compare and merge elements from left and right halves into the original list
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            arr[k] = left_half[i]
            i += 1
        else:
            arr[k] = right_half[j]
            j += 1
        k += 1
    
    # Check if there are any remaining elements in left_half or right_half
    while i < len(left_half):
        arr[k] = left_half[i]
        i += 1
        k += 1
    
    while j < len(right_half):
        arr[k] = right_half[j]
        j += 1
        k += 1

# Input list of numbers
input_list = [34, 10, 32, 15, 62, 17]

# Call the merge_sort function to sort the list
merge_sort(input_list)

# Display the sorted list
print("Sorted List:")
for num in input_list:
    print(num, end=" ")
```

# 6.Write a Python script to perform the following operations on a singly linked list

a. Create a list

b. Find the smallest element from the list

C. Insert an element if it is not a duplicate element

d. Display the elements in reverse order

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    # a. Create a list
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # b. Find the smallest element from the list
    def find_smallest(self):
        if not self.head:
            return None

        smallest = self.head.data
        current = self.head.next
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest

    # c. Insert an element if it is not a duplicate element
    def insert_if_not_duplicate(self, data):
        if not self.head:
            self.head = Node(data)
            return True

        current = self.head
        while current:
            if current.data == data:
                return False  # Duplicate element, not inserted
            if not current.next:
                current.next = Node(data)
                return True
            current = current.next

    # d. Display the elements in reverse order
    def display_reverse(self):
        if not self.head:
            return []

        stack = []
        current = self.head
        while current:
            stack.append(current.data)
            current = current.next

        reversed_list = []
        while stack:
            reversed_list.append(stack.pop())

        return reversed_list


# Example usage:
if __name__ == "__main__":
    linked_list = LinkedList()

    # Create a list
    linked_list.append(20)
    linked_list.append(30)
    linked_list.append(40)
    linked_list.append(50)
    linked_list.append(10)

    
    smallest = linked_list.find_smallest()   # Find the smallest element 
    print("Smallest element:", smallest)

    
    inserted = linked_list.insert_if_not_duplicate(7)    # Insert an element if it's not a duplicate
    if inserted:
        print("Element 7 inserted")
    else:
        print("Element 7 is a duplicate, not inserted")

    
    reversed_elements = linked_list.display_reverse()        # Display elements in reverse order
    print("Elements in reverse order:", reversed_elements)
```

# 7.Write a python program to implement the various operations for Stack ADT

i.) Push ii.) Pop iii.) Display

```python
class Stack:
    def __init__(self):
        self.items = []

    # i.) Push
    def push(self, item):
        self.items.append(item)

    # ii.) Pop
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    # iii.) Display
    def display(self):
        return self.items

    def is_empty(self):
        return len(self.items) == 0


# Example usage:
if __name__ == "__main__":
    stack = Stack()

    # Push elements onto the stack
    stack.push(10)
    stack.push(20)
    stack.push(30)

    
    print("Stack after pushing elements:", stack.display())         # Display the stack

    
    popped_item = stack.pop()                       # Pop elements from the stack
    print("Popped item:", popped_item)

    
    print("Stack after popping one element:", stack.display())     # Display the stack after popping

    # Check if the stack is empty
    if stack.is_empty():
        print("Stack is empty")
    else:
        print("Stack is not empty")
```

# 8.Write a python script to implement the various operations for Queue ADT 
i.) Insert ii.) Delete iii.) Display.

```python
class Queue:
    def __init__(self):
        self.items = []

    # i.) Insert (enqueue)
    def insert(self, item):
        self.items.append(item)

    # ii.) Delete (dequeue)
    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return "Queue is empty"

    # iii.) Display
    def display(self):
        return self.items

    def is_empty(self):
        return len(self.items) == 0


# Example usage:
if __name__ == "__main__":
    queue = Queue()

    # Insert elements into the queue
    queue.insert(10)
    queue.insert(20)
    queue.insert(30)

    
    print("Queue after inserting elements:", queue.display())      # Display the queue

    
    deleted_item = queue.delete()          # Delete elements from the queue
    print("Deleted item:", deleted_item)

  
    print("Queue after deleting one element:", queue.display())      # Display the queue after deleting

    # Check if the queue is empty
    if queue.is_empty():
        print("Queue is empty")
    else:
        print("Queue is not empty")
```

# 9. Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack

a. A/B^C+D*E-F*G

b. (B^2-4 A*C)^(1/2) (100)

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(expression):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    output = []
    stack = Stack()

    for char in expression:
        if char.isalnum():
            output.append(char)
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while stack.peek() != '(' and not stack.is_empty():
                output.append(stack.pop())
            if not stack.is_empty() and stack.peek() != '(':
                return "Invalid Expression"
            else:
                stack.pop()
        else:
            while (not stack.is_empty() and
                   stack.peek() != '(' and
                   precedence[char] <= precedence.get(stack.peek(), 0)):
                output.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        if stack.peek() == '(':
            return "Invalid Expression"
        output.append(stack.pop())

    return ''.join(output)


# Test cases
expression1 = "A/B^C+D*E-F*G"
expression2 = "(B^2-4*A*C)^(1/2)*(100)"

postfix1 = infix_to_postfix(expression1)
postfix2 = infix_to_postfix(expression2)

print("Infix Expression 1:", expression1)
print("Postfix Expression 1:", postfix1)

print("Infix Expression 2:", expression2)
print("Postfix Expression 2:", postfix2)
```
