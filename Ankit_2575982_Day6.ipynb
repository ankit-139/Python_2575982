# 1.Write a python program to create a base class "Shape" with methods to calculate area and perimeter. Then, create derived classes "Circle" and "Rectangle" that inherit from the base class and calculate their respective areas and perimeters. Demonstrate their usage in a program.

You are developing an online quiz application where users can take quizzes on various topics and receive scores.

1. Create a class for quizzes and questions.

2. Implement a scoring system that calculates the user's score on a quiz.

3. How would you store and retrieve user progress, including quiz history and scores?

```python
import math

class Shape:
    def calculate_area(self):
        pass

    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius**2

    def calculate_perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)


circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle Area: {circle.calculate_area()}")
print(f"Circle Perimeter: {circle.calculate_perimeter()}")

print(f"Rectangle Area: {rectangle.calculate_area()}")
print(f"Rectangle Perimeter: {rectangle.calculate_perimeter()}")
```

```python
class Question:
    def __init__(self, text, options, correct_option):
        self.text = text
        self.options = options
        self.correct_option = correct_option

    def is_correct(self, user_answer):
        return user_answer == self.correct_option


class Quiz:
    def __init__(self, title, questions):
        self.title = title
        self.questions = questions

    def take_quiz(self):
        score = 0
        total_questions = len(self.questions)

        for question in self.questions:
            print(question.text)
            for i, option in enumerate(question.options, start=1):
                print(f"{i}. {option}")

            user_answer = int(input("Enter the number of your answer: ")) - 1

            if question.is_correct(user_answer):
                print("Correct!\n")
                score += 1
            else:
                print("Incorrect.\n")

        return score


# Example usage:

# Create questions
q1 = Question("What is the name of Chandrayaan-3 lander?", ["Vikram", "Pragyan", "Aditya"], 0)
q2 = Question("Where did G20 Summit held in 2023??", ["India", "Germany", "America"], 0)
q3 = Question("Which country won t20 world cup 2022?", ["Pakistan", "England", "Australia"], 1)

# Create a quiz
quiz = Quiz("General Knowledge Quiz", [q1, q2, q3])

# Take the quiz and calculate the score
user_score = quiz.take_quiz()
print(f"Your score: {user_score} out of {len(quiz.questions)}")

# Store and retrieve user progress, including quiz history and scores:
user_progress = {
    "username": "example_user",
    "quizzes_taken": [{"quiz_title": quiz.title, "score": user_score}],
}

# To retrieve the user's progress:
print(f"User: {user_progress['username']}")
for quiz_result in user_progress["quizzes_taken"]:
    print(f"Quiz: {quiz_result['quiz_title']}, Score: {quiz_result['score']}")
```

# 2.Write a python script to create a class "Person" with private attributes for age and name. Implement a method to calculate a person's eligibility for voting based on their age. Ensure that age cannot be accessed directly but only through a getter method.

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute for name
        self.__age = age    # Private attribute for age

    
    @property                # Getter method for age
    def age(self):
        return self.__age

    # Method to calculate eligibility for voting
    def is_eligible_to_vote(self):
        if self.__age >= 18:
            return True
        else:
            return False

    # Getter method for name (optional)
    @property
    def name(self):
        return self.__name


person = Person("Ankit", 25) 

# Access age using the getter method
print(f"{person.name}'s Age: {person.age}")

# Check eligibility for voting
if person.is_eligible_to_vote():
    print(f"{person.name} is eligible to vote.")
else:
    print(f"{person.name} is not eligible to vote.")
```

# 4. Employee Managment System.

```python
class Employee:
    def __init__(self, name, employee_id):
        self.__name = name  # Private attribute for name
        self.__employee_id = employee_id  # Private attribute for employee ID
        self.__salary = 0  # Initialize salary to 0

    # Getter method for name
    def get_name(self):
        return self.__name

    # Getter method for employee ID
    def get_employee_id(self):
        return self.__employee_id

    # Setter method for salary
    def set_salary(self, salary):
        self.__salary = salary

    # Calculate salary (to be overridden by subclasses)
    def calculate_salary(self):
        pass

# Subclass for Full Time Employee
class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, annual_salary):
        super().__init__(name, employee_id)
        self.__annual_salary = annual_salary  # Private attribute for annual salary

    # Override the calculate_salary method for Full Time Employee
    def calculate_salary(self):
        return self.__annual_salary

# Subclass for Part Time Employee
class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, hourly_rate, hours_worked):
        super().__init__(name, employee_id)
        self.__hourly_rate = hourly_rate  # Private attribute for hourly rate
        self.__hours_worked = hours_worked  # Private attribute for hours worked

    # Override the calculate_salary method for Part Time Employee
    def calculate_salary(self):
        return self.__hourly_rate * self.__hours_worked

# Demonstrate polymorphism by creating instances of both FullTimeEmployee and PartTimeEmployee
full_time_employee = FullTimeEmployee("Ankit Patel", "EMP01", 90000)
part_time_employee = PartTimeEmployee("John Matt", "EMP02", 20, 25)

# Set salaries for both employees
full_time_employee.set_salary(full_time_employee.calculate_salary())
part_time_employee.set_salary(part_time_employee.calculate_salary())

# Display employee information
print("Full Time Employee Information:")
print(f"Name: {full_time_employee.get_name()}")
print(f"Employee ID: {full_time_employee.get_employee_id()}")
print(f"Salary: ${full_time_employee.calculate_salary()}")

print("\nPart Time Employee Information:")
print(f"Name: {part_time_employee.get_name()}")
print(f"Employee ID: {part_time_employee.get_employee_id()}")
print(f"Salary: ${part_time_employee.calculate_salary()}")
```

# 3. Banking System

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        """
        Constructor for the BankAccount class.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): The initial balance. Defaults to 0.0.
        """
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        """
        Deposit money into the account.

        Args:
            amount (float): The amount to deposit.

        Returns:
            float: The new balance after the deposit.
        """
        if amount > 0:
            self.balance += amount
        return self.balance

    def withdraw(self, amount):
        """
        Withdraw money from the account.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            float: The new balance after the withdrawal, or -1 if there are insufficient funds.
        """
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return -1

# Usage example:
if __name__ == "__main__":
    # Create a BankAccount instance
    account = BankAccount("139", "Ankit Patel", 10000.0)

    # Deposit money
    account.deposit(500.0)
    print(f"New balance after deposit: {account.balance}")

    # Withdraw money
    new_balance = account.withdraw(300.0)
    if new_balance != -1:
        print(f"New balance after withdrawal: {new_balance}")
    else:
        print("Insufficient funds for withdrawal.")
```

```python
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, interest_rate=0.0):
        """
        Constructor for the SavingsAccount class.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): The initial balance. Defaults to 0.0.
            interest_rate (float, optional): The interest rate. Defaults to 0.0.
        """
        super().__init__(account_number, account_holder_name, balance)
        self.interest_rate = interest_rate

    def calculate_interest(self):
        """
        Calculate and add interest to the account based on the interest rate.

        Returns:
            float: The new balance after adding interest.
        """
        interest = self.balance * (self.interest_rate / 100)
        self.balance += interest
        return self.balance

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, overdraft_limit=0.0):
        """
        Constructor for the CheckingAccount class.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): The initial balance. Defaults to 0.0.
            overdraft_limit (float, optional): The overdraft limit. Defaults to 0.0.
        """
        super().__init__(account_number, account_holder_name, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        """
        Withdraw money from the account with overdraft limit.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            float: The new balance after the withdrawal, or -1 if the withdrawal exceeds the overdraft limit.
        """
        if self.balance + self.overdraft_limit >= amount:
            self.balance -= amount
            return self.balance
        else:
            return -1
```

# 5. Library Managment System.

```python
import datetime

class Book:
    def __init__(self, title, author, publication_year):
        self.__title = title
        self.__author = author
        self.__publication_year = publication_year
        self.__is_checked_out = False
        self.__due_date = None

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def get_publication_year(self):
        return self.__publication_year

    def is_checked_out(self):
        return self.__is_checked_out

    def check_out(self, patron, days_due=14):
        if not self.__is_checked_out:
            self.__is_checked_out = True
            today = datetime.date.today()
            self.__due_date = today + datetime.timedelta(days=days_due)
            return f"{self.__title} has been checked out by {patron.get_name()}. Due date: {self.__due_date}"
        else:
            return f"{self.__title} is already checked out."

    def return_book(self):
        if self.__is_checked_out:
            self.__is_checked_out = False
            return f"{self.__title} has been returned."
        else:
            return f"{self.__title} is not checked out."

    def is_overdue(self):
        if self.__is_checked_out and self.__due_date < datetime.date.today():
            return True
        return False

class FictionBook(Book):
    def __init__(self, title, author, publication_year, genre):
        super().__init__(title, author, publication_year)
        self.__genre = genre

    def get_genre(self):
        return self.__genre

class NonFictionBook(Book):
    def __init__(self, title, author, publication_year, topic):
        super().__init__(title, author, publication_year)
        self.__topic = topic

    def get_topic(self):
        return self.__topic

class Patron:
    def __init__(self, name, email):
        self.__name = name
        self.__email = email
        self.__checked_out_books = []

    def get_name(self):
        return self.__name

    def get_email(self):
        return self.__email

    def check_out_book(self, book):
        if book.check_out(self) == f"{book.get_title()} has been checked out by {self.__name}. Due date: {book.is_overdue()}":
            self.__checked_out_books.append(book)
            return f"{self.__name} checked out {book.get_title()}"
        else:
            return f"{self.__name} could not check out {book.get_title()}"

    def return_book(self, book):
        result = book.return_book()
        if result == f"{book.get_title()} has been returned.":
            self.__checked_out_books.remove(book)
        return result

class Transaction:
    def __init__(self, book, patron, transaction_type):
        self.__book = book
        self.__patron = patron
        self.__transaction_type = transaction_type
        self.__date = datetime.date.today()

    def get_book(self):
        return self.__book

    def get_patron(self):
        return self.__patron

    def get_transaction_type(self):
        return self.__transaction_type

    def get_date(self):
        return self.__date

def check_overdue_books(library):
    overdue_books = []
    for book in library:
        if book.is_overdue():
            overdue_books.append(book)
    return overdue_books
```
